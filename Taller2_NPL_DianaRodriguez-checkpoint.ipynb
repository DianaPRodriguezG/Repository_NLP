
- Responder: ¿Cuál archivo tiene el mayor número de palabras?


```python
#Inicializar variables
nroPalabrasMayor = 0
PoemaDef=''
#Recorre cada una de la raiz, directorios, y archivos
for raiz, dirs, archivos in os.walk("Downloads/archivos/poemas"):
    for nombre_archivo in archivos:
        with open("Downloads/archivos/poemas/" + nombre_archivo) as contenido_poema:
            nroLineas = 0
            nroPalabras = 0
            for linea in contenido_poema:
                nroLineas += 1
                palabrasLinea = linea.split(None)
                nroPalabras = nroPalabras + len(palabrasLinea)
            if(nroPalabrasMayor<nroPalabras):
                nroPalabrasMayor=nroPalabras
                PoemaDef=nombre_archivo
print("\n Respuesta => El poema con mayor número de palabras es: '" + PoemaDef + "' el cual tiene " + str(nroPalabrasMayor) + " palabras. \n")
```

## Punto 2

- Hacer Web Scraping de 10 biografías en Wikipedia (en búcle)
- Obtener el **encabezado** de cada biografía
- Obtener todos los **contenidos** y **etiquetas de título** asociados a los **links** del primer párrafo


```python
#Abrir archivo txt
import bs4 as bs
import urllib.request as url
import os
for raiz, dirs, archivos in os.walk("Downloads/Biografias"):
    for nombre_archivo in archivos:
        with open("Downloads/Biografias/" + nombre_archivo) as contenidoBio:
            for linea in contenidoBio:
                link = linea
                #Descagar pagina web en Python
                request = url.urlopen(link)
                fuente = request.read()
                request.close()
                #Procesar pagina web como HTML
                soup = bs.BeautifulSoup(fuente, 'html.parser')
                #Extraer encabezado
                encabezado = soup.find('h1').contents
                print("Encabezado:")
                print(encabezado)
                print()
                #Extraer el contenido y etiquetas de título asociados a los links 
                i = 0
                for primerParrafo in soup.select('div > p'):
                    if len(primerParrafo.contents) > 1 and i < 1: 
                        i += 1
                        for enlaces in primerParrafo.find_all('a'):                
                                print(enlaces)
                                print("Contenido:")
                                print(enlaces.contents)
                                print("Título:")
                                if (enlaces.get('title') != None):
                                    print(enlaces.get('title') + "\n")
                                else:
                                    print('Sin título'+ "\n")
                        print('*'*100)
```
